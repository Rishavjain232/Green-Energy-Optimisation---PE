```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot
```

```python

input1=pd.read_csv('RenTotalEnergy.csv')
input2=pd.read_csv('RenTotalEnergy_nosched.csv')
x1=input1.as_matrix()
x2=input2.as_matrix()
y=np.arange(1,31,1)
```

```python
fig, ax = pyplot.subplots(1)
```

```python
pyplot.plot(y,x1[:,[0]].reshape(-1),'r',label='Load Dist. based on Predicted Ren. Energy')
pyplot.plot(y,x2[:,[0]].reshape(-1),'g',label='Randomised Load Distribution')
# pyplot.plot(y,x1[:,[1]].reshape(-1),'b',label='Randomised Load Distribution')
# pyplot.plot(y,x2[:,[1]].reshape(-1),'y',label='Randomised Load Distribution')
pyplot.xlabel('Days',fontsize=14)
pyplot.ylabel('Renpercent',fontsize=14)
pyplot.title('Renpercent for DC1',fontsize=17)
pyplot.figure(figsize=(30,15))
# pyplot.legend()
pyplot.show()
```
