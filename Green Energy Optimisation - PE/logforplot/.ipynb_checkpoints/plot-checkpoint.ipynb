```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot
```

```python

input=pd.read_csv('RenPercents.csv')
input2=pd.read_csv('RenPercents_nosched.csv')
x1=input.as_matrix()
x2=input2.as_matrix()
y=np.arange(1,31,1)
```

```python
fig, ax = pyplot.subplots(1)
# x[:,[2]]
```

```python
#pyplot.plot(y,x1[1,].as_matrix(),'r',label='DC1')
pyplot.plot(y,x1[:,[0]].reshape(-1),'r',label='Load Dist. based on Predicted Ren. Energy')
pyplot.plot(y,x2[:,[0]].reshape(-1),'g',label='Randomised Load Distribution')
pyplot.xlabel('Days',fontsize=14)
pyplot.ylabel('Renpercent',fontsize=14)
pyplot.title('Renpercent for DC1',fontsize=17)
pyplot.legend()
pyplot.show()



```
