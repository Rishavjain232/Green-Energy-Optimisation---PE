```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot
```

```python
input=pd.read_csv('RenTotalEnergy.csv')
x1=input.as_matrix()
y=np.arange(1,31,1)
ans=x1[:,[1]]/x1[:,[0]]
ans
```

```python
pyplot.plot(y,x1[:,[0]].reshape(-1),'r',label='Total_energy')
pyplot.plot(y,x1[:,[1]].reshape(-1),'g',label='Renewable_energy')
pyplot.xlabel('Days',fontsize=14)
pyplot.ylabel('Kilowat',fontsize=14)
pyplot.title('Relative Energy conpution for DC1',fontsize=17)
pyplot.legend()
pyplot.show()
```
