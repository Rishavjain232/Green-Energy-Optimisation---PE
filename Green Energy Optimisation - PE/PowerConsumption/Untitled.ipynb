```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot
```

```python
x1=pd.read_csv('JobLoad_1_1.csv')
x2=pd.read_csv('JobLoad_1_2.csv')
x3=pd.read_csv('JobLoad_1_3.csv')
x4=pd.read_csv('JobLoad_1_4.csv')

pred1=pd.read_csv('energy_pred_file1')
pred2=pd.read_csv('energy_pred_file2')
pred3=pd.read_csv('energy_pred_file3')
pred4=pd.read_csv('energy_pred_file4')

act1=pd.read_csv('energy_act_file1',header=None)
act2=pd.read_csv('energy_act_file2',header=None)
act3=pd.read_csv('energy_act_file3',header=None)
act4=pd.read_csv('energy_act_file4',header=None)

y=np.arange(1,25,1)
x=x1.as_matrix()
y
# x.reshape(1,24)
```

```python
fig, ax = pyplot.subplots(1)
pyplot.plot(y,x1.as_matrix(),'r',label='DC1')
pyplot.plot(y,x2.as_matrix(),'b',label='DC2')
pyplot.plot(y,x3.as_matrix(),'g',label='DC3')
pyplot.plot(y,x4.as_matrix(),'c',label='DC4')
# pyplot.plot(y,pred1)
#pyplot.plot(data.index.values[size:len(X)],predictions, color='red')
#fig.autofmt_xdate()
pyplot.title('Jobs scheduled on all Datacenters',fontsize=17)
pyplot.xlabel('Hour',fontsize=14)
pyplot.ylabel('No. of Jobs',fontsize=14)
pyplot.legend()
pyplot.show()

print('\n')
pyplot.plot(y,act1,'r',label='DC1')
pyplot.plot(y,act2,'b',label='DC2')
pyplot.plot(y,act3,'g',label='DC3')
pyplot.plot(y,act4,'c',label='DC4')
# pyplot.plot(y,pred1)
#pyplot.plot(data.index.values[size:len(X)],predictions, color='red')
#fig.autofmt_xdate()
pyplot.title('Actual Renewable Energy Generation per hour',fontsize=17)
pyplot.xlabel('Hour',fontsize=14)
pyplot.ylabel('Renewable energy (W)',fontsize=14)
pyplot.legend()
pyplot.show()

print('\n')
pyplot.plot(y,pred1,'r',label='DC1')
pyplot.plot(y,pred2,'b',label='DC2')
pyplot.plot(y,pred3,'g',label='DC3')
pyplot.plot(y,pred4,'c',label='DC4')
# pyplot.plot(y,pred1)
#pyplot.plot(data.index.values[size:len(X)],predictions, color='red')
#fig.autofmt_xdate()
pyplot.title('Predicted Renewable Energy Generation per hour',fontsize=17)
pyplot.xlabel('Hour',fontsize=14)
pyplot.ylabel('Renewable energy (W)',fontsize=14)
pyplot.legend()
pyplot.show()
```
