```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot
```

```python
daynum = "19"
x1=pd.read_csv('./PowerConsumption/JobLoad_'+daynum+'_1.csv')
x2=pd.read_csv('./PowerConsumption/JobLoad_'+daynum+'_2.csv')
x3=pd.read_csv('./PowerConsumption/JobLoad_'+daynum+'_3.csv')
x4=pd.read_csv('./PowerConsumption/JobLoad_'+daynum+'_4.csv')

pred1=pd.read_csv('./PredEnergy/energy_pred_file_'+daynum+'_1',header=None)
pred2=pd.read_csv('./PredEnergy/energy_pred_file_'+daynum+'_2',header=None)
pred3=pd.read_csv('./PredEnergy/energy_pred_file_'+daynum+'_3',header=None)
pred4=pd.read_csv('./PredEnergy/energy_pred_file_'+daynum+'_4',header=None)

act1=pd.read_csv('./ActualEnergy/energy_act_file_'+daynum+'_1',header=None)
act2=pd.read_csv('./ActualEnergy/energy_act_file_'+daynum+'_2',header=None)
act3=pd.read_csv('./ActualEnergy/energy_act_file_'+daynum+'_3',header=None)
act4=pd.read_csv('./ActualEnergy/energy_act_file_'+daynum+'_4',header=None)

y=np.arange(1,25,1)
# x=x1.as_matrix()
```

```python
def plotGraph(X1,X2,Y,dc):
#     fig = pyplot.figure(figsize=(10,5))
#     ax1 = fig.add_subplot(111)
    fig, ax1 = pyplot.subplots(1)

    ax1.plot(Y,X1.as_matrix(),'b',label='Load')
    ax1.set_ylabel("Number of Jobs", color='b',fontsize=14)
    ax1.tick_params('y', colors='b')
    pyplot.legend()

    ax2 = ax1.twinx()
    ax2.plot(Y,X2,'r',label='Predicted Ren. Energy')
    ax2.set_ylabel("Predicted Ren. Energy", color='r',fontsize=14)
    ax2.tick_params('y', colors='r')
    pyplot.title(dc,fontsize=17)
    pyplot.xlabel('Hour',fontsize=14)
    pyplot.legend()

    fig.tight_layout()
    pyplot.savefig("./z_plots/job_ren_1_"+dc+".png")
    pyplot.show()
```

```python
plotGraph(x1,pred1,y,'DC1')
plotGraph(x2,pred2,y,'DC2')
plotGraph(x3,pred3,y,'DC3')
plotGraph(x4,pred4,y,'DC4')
```
